```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
# Reference the file where the CSV is located
instacart_csv_path = "LimitedData.csv"

#Import the data into a Pandas Datafram
instacart_df = pd.read_csv(instacart_csv_path)
instacart_df.head(50)
```

```python
#create a new column with a conditional numpy where statement 
# instacart_df['Organic or Not'] = instacart_df.loc[instacart_df['product_name'].str.contains('Organic')]
instacart_df['Organic or Not'] = instacart_df['product_name'].apply(lambda x : 'Organic' if 'Organic' in x else 'Not Organic')
instacart_df

```

```python
unique_arr = instacart_df["aisle"].unique()
unique_arr
```

```python
#instacart_df['Veg Friendly'] = instacart_df['aisle'].apply(lambda x : 'Vegetarian' if x.split() in ['fresh herbs', 'fresh vegetables', 'fresh fruits'] else 'Non Vegetarian')
instacart_df['Veg Friendly'] = instacart_df['aisle'].apply(lambda x : 'Vegetarian' if x in ['fresh herbs','fresh fruits','fresh vegetables', 'packaged vegetables fruits', 'butter', 'oils vinegars', 'pasta sauce', 'spreads', 'milk', 'canned meals beans', 'eggs',
       'water seltzer sparkling water', 'soy lactosefree',
       'nuts seeds dried fruit', 'packaged cheese',
       'canned jarred vegetables', 'bread', 'specialty cheeses',
       'preserved dips spreads', 'breakfast bakery',
       'fresh dips tapenades', 'tofu meat alternatives',
       'energy granola bars', 'cereal', 'cream', 'dry pasta', 'yogurt', 'packaged produce', 'crackers', 'chips pretzels', 'juice nectars', 'prepared soups salads',
       'doughs gelatins bake mixes', 'fruit vegetable snacks',
       'canned fruit applesauce', 'other creams cheeses', 'soft drinks',
       'trail mix snack mix', 'mint gum', 'cookies cakes', 'tea',
       'packaged poultry', 'frozen vegan vegetarian',
       'frozen breads doughs', 'soup broth bouillon', 'candy chocolate'] else 'Non Veg')
instacart_df
```

```python
instacart_clean = instacart_df.drop(columns=['Organic vs Non Organic', 'Vegan Friendly', 'Food Group'])
instacart_clean.head(45)
```
